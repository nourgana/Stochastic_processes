```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
S0 = 100
r = 0.05
sigma = 0.2
lamb = 0.75
mu = -0.6
delta = 0.25 

T = 1
N = 50 #I 
M = 50
h = T/N
rj = lamb*(np.exp(mu+0.5*delta**2)-1)
t = np.linspace(0,T,N+1,endpoint=True)
```

```python
S = np.zeros((N+1,M))
S[0,:] = S0


z1 = np.random.randn(N+1,M)
z2 = np.random.randn(N+1,M)
y = np.random.poisson(lamb*h,(N+1,M))

for s in range(M):
    
    for i in range(1,N+1):
        
        temp = S[i-1,s] * (np.exp((r-rj-0.5*sigma**2)*h+sigma*np.sqrt(h)*z1[i,s]) + (np.exp(mu+delta*z2[i,s])-1)*y[i,s])
        
        S[i,s] = max(0,temp)
        

plt.figure(figsize=(16,9))
plt.plot(t,S)
plt.title("Schéma d'Euler du processus de saut de Merton")
plt.xlabel("Temps")
plt.grid()
plt.show()
```
